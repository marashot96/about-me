# –ü–µ—Ä–µ–≤–æ–¥ –∞—Ä–∞–±—Å–∫–∏—Ö —á–∏—Å–µ–ª –≤ –∫–∏—Ç–∞–π—Å–∫—É—é —á–∏—Å–ª–æ–≤—É—é –∑–∞–ø–∏—Å—å —Å —É—á–µ—Ç–æ–º —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤


**–¶–µ–ª—å:** –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –∞—Ä–∞–±—Å–∫–æ–µ —á–∏—Å–ª–æ (—Ü–µ–ª–æ–µ –∏–ª–∏ –¥—Ä–æ–±–Ω–æ–µ, –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –∏–ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ) –≤ —á–∏—Å–ª–æ–≤—É—é –∑–∞–ø–∏—Å—å –Ω–∞ –∫–∏—Ç–∞–π—Å–∫–æ–º —è–∑—ã–∫–µ.

**–ê–ª–≥–æ—Ä–∏—Ç–º:**
–ß–∏—Å–ª–æ —Ä–∞–∑–±–∏–≤–∞–µ—Ç—Å—è –Ω–∞:
- —Ü–µ–ª—É—é —á–∞—Å—Ç—å (–æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç—Å—è —Ä–∞–∑—Ä—è–¥–∞–º–∏ `ÂçÉ`, `Áôæ`, `ÂçÅ`, `‰∏Ä` –∏ –∫—Ä—É–ø–Ω—ã–º–∏: `‰∏á`, `‰∫ø`, `ÂÖÜ`);
- –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å (–ø–µ—Ä–µ–≤–æ–¥–∏—Ç—Å—è –ø–æ—Å–∏–º–≤–æ–ª—å–Ω–æ –ø–æ—Å–ª–µ ÁÇπ);
- –ø–µ—Ä–µ–¥ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º–∏ —á–∏—Å–ª–∞–º–∏ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è —Å–∏–º–≤–æ–ª Ë¥ü.

–¢–∞–∫–∂–µ —Ñ—É–Ω–∫—Ü–∏—è —É—á–∏—Ç—ã–≤–∞–µ—Ç —Å–ª–µ–¥—É—é—â–∏–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –∑–∞–ø–∏—Å–∏ –∫–∏—Ç–∞–π—Å–∫–∏—Ö —á–∏—Å–µ–ª:
- –Ω–æ–ª—å;
- —á–∏—Å–ª–∞ 10‚Äì19 –±–µ–∑ "‰∏Ä" –ø–µ—Ä–µ–¥ "ÂçÅ";
- –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞;
- –¥—Ä–æ–±–Ω—ã–µ —á–∞—Å—Ç–∏;
- —É–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –Ω—É–ª–µ–π.

**–ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:**

`print(to_chinese_numeral(-10.000001))` ‚Üí `Ë¥üÂçÅÁÇπÈõ∂Èõ∂Èõ∂Èõ∂Èõ∂‰∏Ä`

**–ü—Ä–æ–≥—Ä–∞–º–º—É –º–æ–∂–Ω–æ –∑–∞–ø—É—Å—Ç–∏—Ç—å –∑–¥–µ—Å—å:**

**–ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º—ã:**

```python
# –°–ª–æ–≤–∞—Ä–∏ –¥–ª—è –±–∞–∑–æ–≤—ã—Ö —Ü–∏—Ñ—Ä –∏ —Ä–∞–∑—Ä—è–¥–æ–≤
ch = {
    '-': 'Ë¥ü',
    0: 'Èõ∂',
    1: '‰∏Ä',
    2: '‰∫å',
    3: '‰∏â',
    4: 'Âõõ',
    5: '‰∫î',
    6: 'ÂÖ≠',
    7: '‰∏É',
    8: 'ÂÖ´',
    9: '‰πù',
    10: 'ÂçÅ',
    100: 'Áôæ',
    1000: 'ÂçÉ',
    10000: '‰∏á',
    '.': 'ÁÇπ'
}

#–ë–ª–æ–∫, –ø–µ—Ä–µ–≤–æ–¥—è—â–∏–π –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –∫–∏—Ç–∞–π—Å–∫—É—é —Å—á–µ—Ç–Ω—É—é —Å–∏—Å—Ç–µ–º—É –±–µ–∑ —É—á–µ—Ç–∞ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–µ–π
def f(numbs):
    res = []
    r = ''
    for i in range(len(numbs[0])):
        if numbs[0][i] != '0':
            r += ch[int(numbs[0][i])] + ch[10**(len(numbs[0]) - i - 1)]
        else:
            r += ch[0]
    if numbs[0] != '0':
        res.append(r[:-1])
    else:
        res.append(ch[0])
    if len(numbs) == 2:
        k = ''
        for i in numbs[1]:
            k += ch[int(i)]
        res.append(k)
    return res

def to_chinese_numeral(n):
    if n == 0:
        return ch[0]

    resu = ''
    negative = n < 0
    if negative:
        resu += ch['-']
        n = abs(n)

    if n == 10:
        return resu + ch[10]
    elif 10 < n < 20 and int(n) == n:
        return resu + ch[10] + ch[int(n) % 10]

    str_n = str(n)
    if '.' in str_n:
        int_part, frac_part = str_n.split('.')
    else:
        int_part, frac_part = str_n, None

    if int_part == '10' and frac_part:
        short_ten = True
    else:
        short_ten = False

    numbs = [int_part]
    if frac_part:
        numbs.append(frac_part)

    parts = f(numbs)
    integer_part = parts[0]
    while ch[0]*2 in integer_part:
        integer_part = integer_part.replace(ch[0]*2, ch[0])
    if integer_part.endswith(ch[0]):
        integer_part = integer_part[:-1]

    if int_part == '0':
        resu += ch[0]
    elif int_part == '10' and not frac_part:
        resu += ch[10]
    elif int_part == '10' and frac_part:
        resu += ch[10]
    elif int_part == '1' and frac_part is None and 10 <= n < 20:
        resu += ch[10] + ch[int(n) % 10]
    else:
        resu += integer_part

    if frac_part:
        resu += ch['.'] + parts[1]

    return resu
```

**–ó–∞–ø—É—Å–∫ –ø—Ä–æ–≥—Ä–∞–º–º—ã:**

```python
print(to_chinese_numeral(0))              
print(to_chinese_numeral(10))             
print(to_chinese_numeral(14))             
print(to_chinese_numeral(101))            
print(to_chinese_numeral(1001))           
print(to_chinese_numeral(10000))          
print(to_chinese_numeral(61800.5))        
print(to_chinese_numeral(0.5))            
print(to_chinese_numeral(-10.000001))     
print(to_chinese_numeral(-18))            
```
